```python
#Filtrando columns dado um determinado valo(Numerico, string, etc)
dados.loc[dados['Nationality'] == 'Brazil']
```

```python
#Sorteando os valores das columns
dados.sort_values("Name")
```

```python
#Sorteando por idade
dados.sort_values("Age")
```

```python
#Sorteando do maior pro menor
dados.sort_values("Age", ascending = False)
```

```python
#Adicionando uma nova coluna
dados['Total'] = dados['Acceleration'] + dados['Agility'] + dados['Reactions']
dados
```

```python
#Fazendo um filtro na tabela, para mostrar minha nova column e o nome
dadosNew = dados[["Name","Total"]]
dadosNew
```

```python
#Ordenando para mostrar minha nova column e o nome do maior para menor
dadosNew = dados[["Name","Total"]]
dadosNew.sort_values("Total", ascending = False)
```

```python
#Exportando a tabela para um csv
dadosNew.to_csv("dados1.csv", index = False)
```

```python
#Exportando para um excel
dadosNew.to_excel("dadosExecel.xlsx", index = False)
```
